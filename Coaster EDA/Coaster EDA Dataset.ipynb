## Data Preparations
- Dropping irrelevant columns and rows
- Identifying duplicated columns
- Renaming columns
- Feature Creation

```python
df.head()
```

```python
# Choosing relevant Columns
df = df[
    [
        "coaster_name",  #'Length', 'Speed',
        "Location",
        "Status",
        #'Opening date','Type',
        "Manufacturer",
        #'Height restriction', 'Model', 'Height',
        #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
        #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
        #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
        #'Track layout', 'Fastrack available', 'Soft opening date.1',
        #'Closing date',
        #'Opened', 'Replaced by', 'Website',
        #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
        #'Single rider line available', 'Restraint Style',
        #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
        "year_introduced",
        "latitude",
        "longitude",
        "Type_Main",
        "opening_date_clean",
        #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
        "speed_mph",
        #'height_value',
        "height_unit",
        "height_ft",
        "Inversions_clean",
        "Gforce_clean",
    ]
].copy()
```

```python
# Dropping more column
df.drop(['height_unit'], axis=1, inplace=True)
```

```python
df.dtypes
```

```python
# Convert opening_date_clean into Date
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.dtypes
```

```python
# Renaming and make our columns into Proper Case
df.columns = df.columns.str.title()
df.rename(
    columns={
        "Gforce_Clean": "Gforce",
        "Inversions_Clean": "Inversions",
        "Opening_Date_Clean": "Opening_Date",
    },inplace=True
)
```

```python
df.columns
```

```python
# Checking for missing value
df.isna().sum()
```

```python
# Checking for duplicate value in dataframe
df.loc[df.duplicated()]
```

```python
# Checking for duplicated values in Coaster_Name
df.loc[df.duplicated(subset=['Coaster_Name'])]
```

```python
# Suspect the data that duplicated by Coaster_Name
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
# Choose a data that not duplpicated and reset the index
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])].reset_index(drop=True)
```

# Ask a Question
-  What are the locations with the fastest roller coasters? (minimum of 10)

```python
df.head()
```

```python
df.query('Location != "Other"')\
    .groupby(by='Location')['Speed_Mph']\
    .agg(['mean', 'count'])\
    .query('count >= 10')\
    .sort_values(by='mean')['mean']\
    .plot(kind='barh', title='Average Coaster Speed by Location', xlabel='Average Coaster Speed')
plt.show()
```
